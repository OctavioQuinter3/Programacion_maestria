# Instrucciones de la actividad

Creación de un arreglo y operaciones básicas.

Crea un arreglo unidimensional de 20 elementos con valores aleatorios entre 1 y 100. Luego, realiza las siguientes operaciones:

Encuentra el valor máximo del arreglo.

Encuentra el valor mínimo del arreglo.

Calcula la media del arreglo.

Suma 10 a cada elemento del arreglo.

#Inicio de codigo

Importamos la libreria a utilizar, en este caso Numpy para el manejo de los vectores y arreglos.

````python
import numpy as np
````

Creamos nuestro arreglo de 20 elementos aleatorios entre 1 y 100, para esto usaremos lo siguiente:

```Python
arreglo = np.random.randint(1, 101, 20)
```

donde arreglo es nuestra variable a crear, np es numpy, random.randint para la creacion de numeros random enteros, (1,101,20), 1 siendo el inicio, 101 -1 es el maximo 100, y 20 la cantidad de numeros.

````python
arreglo = np.random.randint(1, 101, 20)
print(arreglo)
````

Realizamos el siguiente codigo para lo que se nos pide

#Valor maximo del arreglo

````python
print(arreglo.max())
````

#Valor minimo

````python
print(arreglo.min())
````

#Media del arreglo

````python
print(arreglo.mean())
````

#Suma de 10 a cada elemento del arreglo

````python
arreglo_sum = arreglo + 10
print(arreglo_sum)
````

En este ultimo caso es cuando mas hacemos uso de una propiedad de numpy para no realizar un ciclo for para sumar a cada elemento, con numpy lo podemos realizar directamente.
