# Instrucciones de la actividad

Uso de máscaras booleanas.

Crea un arreglo unidimensional con 15 números aleatorios entre 1 y 50. Luego:

Selecciona los elementos mayores que 25 utilizando una máscara booleana.

Cuenta cuántos elementos son mayores que 25.

#Inicio de codigo

```python
import numpy as np
```

Importamos la libreria a utilizar, en este caso Numpy para el manejo de matrices.

```python
arreglo = np.random.randint(1, 51, 15)

print(arreglo)
```

Como observamos en las practicas anteriores realizamos lo aprendido para la creacion del arreglo random, poniendo minimo 1 a maximo 50 y 15 elementos.

#Condicion para mascara

```python
mascara = arreglo > 25
print(mascara)
```

Como observamos el arreglo de mascara nos da true false, esto hace que por posicion sepamos cual cumple con la condicional de mayor a 25

#Uso de mascara

```python
seleccion_25 = arreglo[mascara]
print(seleccion_25)
```

Aplicamos la mascara a nuestro arreglo original y obtenmos el resultado

#Conteo de elementos en arreglo

```python
print(len(seleccion_25))
```

Para contar cuantos elementos estan en el arreglo resultante, aplicamos len
