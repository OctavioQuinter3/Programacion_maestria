#Inicio de actividad

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

Cargamos el dataset de pinguinos de seaborn.

````python
df = sns.load_dataset("penguins")
````

##Descripcion de los datos

Aqui usamos info para saber si tenemos valores nulos y conocer mas sobre las columnas que tiene el archivo.

````python
df.info()
df.describe()
````

Como podemos ver en info, vemos que si tenemos valores nulos, yo usare
```python
.dropna()
```
para limpiar los datos nulos.

````python
df = df.dropna()
````

Ahora creamos la figura 2,2, y realizaremos cada uno de los diferentes tipos de plots con seaborn.

````python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

sns.histplot(df["flipper_length_mm"], ax=ax[0,0], color="blue")
ax[0,0].set_title("Distribucion de Longitud de Aletas")
ax[0,0].set_xlabel("Longitud de las aletas")
ax[0,0].set_ylabel("Conteo")

sns.boxplot(x="species", y="body_mass_g", data=df, ax=ax[0,1], color="salmon")
ax[0,1].set_title("Comparacion de Masa Corporal por Especie")
ax[0,1].set_xlabel("Especie")
ax[0,1].set_ylabel("Masa Corporal")

sns.scatterplot(x="body_mass_g", y="flipper_length_mm", data=df, ax=ax[1,0], color="lightgreen")
ax[1,0].set_title("Relacion entre Masa Corporal y Longitud de Aletas")
ax[1,0].set_xlabel("Masa Corporal")
ax[1,0].set_ylabel("Longitud de las aletas")

sns.violinplot(x="species", y="body_mass_g", data=df, ax=ax[1,1], color="coral")
ax[1,1].set_title("Distribucion de Masa Corporal por Especie")
ax[1,1].set_xlabel("Especie")
ax[1,1].set_ylabel("Masa Corporal")

plt.tight_layout()
plt.show()
````
