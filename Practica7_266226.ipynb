#Inicio de actividad

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

#Importacion de dataset

````python
df = sns.load_dataset('iris')
````

#Visualizacion de los datos

````python
df.head(10)
````

Informacion del dataset y valores de estadistica

````python
df.info()
df.describe()
````

#Graficacion

##Graficos de barras

###Calculo de promedios de petal length y width por species

````python
promedios = df.groupby('species')[['petal_length', 'petal_width']].mean()
````

###Grafico

````python
promedios.plot(kind='bar')
plt.xlabel('Species')
plt.ylabel('Promedio')
plt.title('Promedios de petal_length y petal_width por species')
plt.show()
````

##Histograma


###Grafico

````python
plt.figure(figsize=(8, 6))
plt.hist(df['sepal_length'], bins=20, color='red', edgecolor='black')
plt.xlabel('Sepal Length')
plt.ylabel('Frecuencia')
plt.title('Distribucion de Sepal Length')
plt.show()
````

##Scatter plot

###Explicacion

Usare un ciclo for donde por cada especie hare un filtrado, quiero decir lo siguiente

En
```python
for especie in df['species'].unique():
```
En el for lo que estoy diciendo es que realice una iteracion por cada especie unica, se tiene 3 especies: setosa, versicolor y virginica.

```python
subset = df[df['species'] == especie]
```
Esto es para tener solo las filas del DataFrame donde la especie sea la actual, ejemplo Si especie = 'setosa' , subset contiene solo flores setosa.

```python
plt.scatter(subset['petal_length'], subset['petal_width'], label=especie)
```
Este es el que grafica los valores en funcion de la especie.

###Grafico

````python
plt.figure(figsize=(7,5))

for especie in df['species'].unique():
    subset = df[df['species'] == especie]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=especie)

plt.title('Relacion entre largo y ancho del petalo')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.legend()
plt.show()
````

##Box plot

###Grafico

````python
df.boxplot(column=['sepal_length', 'sepal_width'], by='species')
plt.show()
````

##Grafico de lineas

###Explicacion

Usare la misma logica de los subset para dividir por especie, pero agregare un reset_index para que las graficas inicien en 0, si no usare esto y usara el index normal, las especies iniciarian en diferentes por lo que no podria comparar directamente.

````python
colores = {
    'setosa': 'blue',
    'versicolor': 'green',
    'virginica': 'red'
}

for especie in df['species'].unique():
    subset = df[df['species'] == especie].reset_index(drop=True)
    plt.plot(
        subset.index,
        subset['petal_length'],
        color=colores[especie],
        label=especie
    )

plt.title('Tendencia del largo del petalo por especie')
plt.xlabel('Index')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
````
