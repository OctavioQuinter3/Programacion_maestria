##Proporcion por genero y clase

Se nos pide obtener la proporcion de supervivencia por genero y clase en el barco, algo importante es que en la columna Suvived es booleana, solamente tenemos 1 cuando sobrevive y 0 cuando no, por lo cual podemos aprovechar esta propiedad, ejemplo [1,0,0,1,1,0] tenemos 6 elementos, si realizamos la suma de los elementos nos da un valor de 3, por lo que si lo dividimos entre el total de elementos tendremos 0.5, esto mismo aplicaremos en todo el data frame.

````python
genero_max, clase_max = genero_clase_proporcion.idxmax()
print("La combinacion que tuvo la tasa mas alta de supervivencia fue: ","Genero:",genero_max,"Clase:",clase_max)
````

#Inicio de ejercicio 3

##Creacion de funcion
Use el tip de hacer una funcion para clasificar si es menor de edad o mayor, ademas de eso me di cuenta que hay valores en los cuales no se sabe la edad y tienen valor NaN, por lo cual podemos usar de pandas.isnan para categorizar como sin informacion y ayudarnos a categorizar.

````python
def clasificar_edad(edad):
    if pd.isna(edad) :
      return 'Sin informacion'
    elif edad < 18:
        return 'Menor de edad'
    else:
        return 'Mayor de edad'
````

##Uso de funcion

Aplicamos la funcion para tener esta categorizacion de menor, mayor.

Yo le puse Edad a la variable, al hacer retrospectiva puede ser que lo mejor hubiera sido poner rango_edad o algo mas descriptivo porque puede confundirse con el Age.

````python
df['Edad'] = df['Age'].apply(clasificar_edad)
df[['Edad','Age']].head(100)
````

#Inicio de Ejercicio 5

##Investigacion de linspace

Linspace es para crear limites equidistantes, esto nos genera n valores equidistantes entre un inicio y un fin.

Donde
``` python
linspace(inicio,fin,n)
```


##Creacion de minimo, maximo para fare

Tendremos el valor minimo o inicio y el valor maximo o fin para poder realizar nuestros limites.

````python
min_fare = df['Fare'].min()
max_fare = df['Fare'].max()
print(min_fare, max_fare)
````

##Uso de linspace

Ya obtuvimos y creamos nuestras variables para inicio, fin que son min_fare y max_fare, n sera 6 para crear 5 intervalos equidistantes, imprimire los intervalos y podemos observar los limites por intervalo.

````python
intervalos = np.linspace(min_fare, max_fare, 6)
print(intervalos)
````

##Creacion de columna Intervalo_Fare

Usaremos pd.cut para asignar a cada uno el rango en el que cae utilizando su variable fare, yo hice los 5 labels para reconocerlo mas rapidamente.

````python
df['Intervalo_Fare'] = pd.cut(df['Fare'], bins=intervalos, labels=['muy bajo', 'bajo', 'medio', 'altop','muy alto'])
df.head(10)
````

##Conteo por intervalo

Contaremos las cantidades por intervalos haciendo uso de valo_counts.

````python
conteo_intervalo = df['Intervalo_Fare'].value_counts()
print(conteo_intervalo)
````

##Proporcion de supervivencia por intervalo

Aqui podemos ver su proporcion de supervivencia por intervalos equidistantes.

````python
proporcion_supervivencia_intervalo = df.groupby('Intervalo_Fare')['Survived'].mean()
print(proporcion_supervivencia_intervalo)
````
