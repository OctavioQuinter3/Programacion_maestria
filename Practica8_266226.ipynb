#Graficacion

##Heatmap

Crearemos la coorelacion pedida de total_bill, tip y size con .coor

```python
correlacion = df[['total_bill','tip','size']].corr()
```

Usaremos seaborn para la creacion del heatmap

```python
sns.heatmap(correlacion, cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Heatmap de correlacion')
plt.show()
```

###Explicacion de la grafica
Podemos ver en la grafica como a medida que tenemos un total bill mas grande tambien tenemos un tip mas alto, en cambio el size con el tip tiene una correlacion mas baja como vemos en el color mas frio.

Yo estoy usando colores calidos y frios para la correlacion.

##Violin

```python
sns.violinplot(x='day', y='tip', data=df, palette='pastel')
plt.title('Distribucion de propinas por dia')
plt.xlabel('Dia')
plt.ylabel('Propinas')
plt.show()
```

###Explicacion de la grafica

Podemos ver que en sabado tenemos propinas mas altas, pero su densidad esta mas distribuida como podemos ver en el largo del violin, el dia domingo tenemos valores arribas en las propinas y vemos datos mas concentrados.

En jueves tenemos datos mas concentrados en valores mas bajos.

El viernes la concentracion es mas alta, su largo del violin no es tan alto como el sabado sin embargo su media esta mas arriba.

El dia de la semana influye en la cantidad que se tendra la propina

##Grafico de dispersion

```python
colores = {
    'Thur': 'black',
    'Fri': 'green',
    'Sat': 'blue',
    'Sun': 'red'
}

sns.scatterplot(x='total_bill', y='tip', hue='day', data=df, palette=colores)
plt.title('Relacion de cuenta con propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

###Explicacion de la grafica

Podemos observar que a mayor total de la cuenta mayor es la propina, pero tambien podemos observar que en fin de semana existen mas mayor total de la cuenta.

##Grafico de barras

```python
sns.barplot(x='sex', y='tip', data=df, palette='pastel')
plt.xlabel('Genero')
plt.ylabel('Propinas')
plt.title('Propinas por genero')
plt.show()
```

###Explicacion de la grafica

En las propinas por genero vemos que el promedio de las propinas es mayor en hombres que en mujeres pero no mucho.
