# Instrucciones de la actividad

Crea dos matrices de 3x3 con números aleatorios entre 1 y 10. Realiza las siguientes operaciones:

Suma ambas matrices.

Resta la segunda matriz de la primera.

Multiplica la primera matriz por 2.

Realiza una multiplicación matricial entre las dos matrices.

#Inicio de codigo

````python
import numpy as np
````

Importamos la libreria a utilizar, en este caso Numpy para el manejo de matrices.

````python
matriz_a = np.random.randint(1, 11,(3, 3))
matriz_b = np.random.randint(1, 11,(3, 3))

print("Matriz A:")
print(matriz_a)
print("\nMatriz B:")
print(matriz_b)
````

Creamos nuestra matriz a y b, utilizamos lo siguiente

```Python
np.random.randint(1, 11,(3, 3))
```
donde random.randint para la creacion de numeros random enteros donde (1,11) es el valor minimo y 11 es el valor maximo quitandole 1 siendo 10 y (3, 3) el tamaño de la matriz

#Suma de matrices

````python
suma = matriz_a + matriz_b
print("Suma de matrices:")
print(suma)
````

#Resta de segunda matriz a primera

````python
resta = matriz_a - matriz_b
print("Resta de matrices:")
print(resta)
````

#Multiplicacion por 2 a matriz A

````python
multiplicacion = matriz_a * 2
print("Multiplicacion por 2 de matriz A:")
print(multiplicacion)
````

Como vimos en este caso es la multiplicaion escalar, a continuacion realizaremos la multiplicacion de matricial que utiliza .dot para realizarlo.

````python
producto_matricial = np.dot(matriz_a, matriz_b)
print("Producto matricial de las matrices:")
print(producto_matricial)
````
